### Now you want to compare the sales of different vehicle types during a recession and a non-recession period
<br>We recommend that you use the functionality of **Seaborn Library** to create this visualization


<details><summary>Click here for a hint</summary>

<p>

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Set Seaborn style
sns.set(style="whitegrid")

# Group data to get average sales by vehicle type and recession status
grouped_df = df.groupby(['Vehicle_Type', 'Recession'])['Automobile_Sales'].mean().reset_index()

# Map recession to readable labels
grouped_df['Recession_Label'] = grouped_df['Recession'].map({0: 'Non-Recession', 1: 'Recession'})

# Set figure size
plt.figure(figsize=(12, 6))

# Create bar plot
sns.barplot(
    data=grouped_df,
    x='Vehicle_Type',
    y='Automobile_Sales',
    hue='Recession_Label',
    palette='Set2'
)

# Add titles and labels
plt.title('Average Automobile Sales by Vehicle Type: Recession vs Non-Recession')
plt.xlabel('Vehicle Type')
plt.ylabel('Average Sales')
plt.legend(title='Period')
plt.xticks(rotation=30)
plt.tight_layout()

# Show plot
plt.show()
````

### TASK 1.5: Develop a Bubble plot for displaying the impact of seasonality on Automobile Sales.
<br>How has seasonality impacted the sales, in which months the sales were high or low? Check it for non-recession years to understand the trend

##### Develop a Bubble plot for displaying Automobile Sales for every month and use Seasonality Weight for representing the size of each bubble<br>
Title this plot as 'Seasonality impact on Automobile Sales'


<details><summary>Click here for a hint</summary>

<p>
     You can create Bubble Chart by calling the scatter()
        <br>Pass the 'Month' and 'Automobile_Sales' to the functions as x and y and then use Seasonality weight for size parameter</p>
</details>


````python
import matplotlib.pyplot as plt

# Filter non-recession data
non_recession_df = df[df['Recession'] == 0]

# Group by Month and calculate total sales and average seasonality weight
seasonality_data = non_recession_df.groupby('Month').agg({
    'Automobile_Sales': 'sum',
    'Seasonality_Weight': 'mean'
}).reset_index()

# Plotting the bubble plot
plt.figure(figsize=(12, 6))
plt.scatter(
    seasonality_data['Month'],
    seasonality_data['Automobile_Sales'],
    s=seasonality_data['Seasonality_Weight'] * 1000,  # Scale the bubble sizes
    color='skyblue',
    edgecolors='black',
    alpha=0.6
)

plt.title('Seasonality Impact on Automobile Sales')
plt.xlabel('Month')
plt.ylabel('Total Automobile Sales')
plt.xticks(seasonality_data['Month'])  # Ensure all months are shown
plt.grid(True)
plt.show()
````
